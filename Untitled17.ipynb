```python


from google import genai
from google.colab import userdata

# Get API key from Colab Secrets
api_key = userdata.get('api_key')

# Create Gemini client
client = genai.Client(api_key=api_key)

# Prompt
prompt = """
You are an AI travel planning team.

Agent 1 (Destination Expert):
Suggest the best destination in the world for a friends trip.

Agent 2 (Budget Planner):
Estimate the total cost for 5 people for 7 days.

Agent 3 (Hotel Expert):
Recommend a hotel category.

Agent 4 (Activity Planner):
Suggest famous tourist activities.

Agent 5 (Travel Manager):
Prepare the final itinerary.

Return the output in the following format:

Destination:
Budget:
Hotel:
Activities:
Travel Plan:
"""

# Generate response
response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents=prompt
)

# Print response
print(response.text)
```
