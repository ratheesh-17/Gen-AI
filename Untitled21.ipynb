```python
# ==========================================================
# HR Dashboard Prompt Engineering Assignment (Single Cell)
# ==========================================================

import json
import pandas as pd

print("="*80)
print("TASK 1: INITIAL PROMPT")
print("="*80)

initial_prompt = """
Generate candidate profiles for an HR dashboard.

Requirements:
1. Generate exactly 5 candidate profiles.
2. Age must be between 21 and 35.
3. Experience must be between 0 and 10 years.
4. Each candidate must have at least 3 skills.
5. Include:
   - CandidateID
   - Name
   - Age
   - ExperienceYears
   - Skills
   - Education
   - Location
6. Output must be provided in:
   - Valid JSON format
   - Table format
"""

print(initial_prompt)

print("\n" + "="*80)
print("TASK 2: FIRST OUTPUT (JSON)")
print("="*80)

candidates = [
    {
        "CandidateID": "C001",
        "Name": "Aarav Sharma",
        "Age": 24,
        "ExperienceYears": 2,
        "Skills": ["Python", "SQL", "Power BI"],
        "Education": "B.Tech AI & DS",
        "Location": "Coimbatore"
    },
    {
        "CandidateID": "C002",
        "Name": "Priya Nair",
        "Age": 27,
        "ExperienceYears": 4,
        "Skills": ["Java", "Spring Boot", "MySQL"],
        "Education": "B.E CSE",
        "Location": "Chennai"
    },
    {
        "CandidateID": "C003",
        "Name": "Rahul Verma",
        "Age": 30,
        "ExperienceYears": 7,
        "Skills": ["AWS", "Docker", "Kubernetes"],
        "Education": "M.Tech IT",
        "Location": "Bengaluru"
    },
    {
        "CandidateID": "C004",
        "Name": "Sneha Iyer",
        "Age": 22,
        "ExperienceYears": 1,
        "Skills": ["React", "JavaScript", "HTML", "CSS"],
        "Education": "B.Tech IT",
        "Location": "Hyderabad"
    },
    {
        "CandidateID": "C005",
        "Name": "Karthik Raj",
        "Age": 35,
        "ExperienceYears": 10,
        "Skills": ["Machine Learning", "Python", "TensorFlow"],
        "Education": "M.Sc Data Science",
        "Location": "Coimbatore"
    }
]

print(json.dumps(candidates, indent=4))

print("\n" + "="*80)
print("TASK 2: TABLE FORMAT")
print("="*80)

df = pd.DataFrame(candidates)
df["Skills"] = df["Skills"].apply(lambda x: ", ".join(x))
print(df.to_string(index=False))

print("\n" + "="*80)
print("TASK 3: OUTPUT VALIDATION")
print("="*80)

validation = []

# Exactly 5 candidates
validation.append(("Exactly 5 Candidates",
                   len(candidates) == 5))

# Age validation
validation.append(("Age between 21 and 35",
                   all(21 <= c["Age"] <= 35 for c in candidates)))

# Experience validation
validation.append(("Experience between 0 and 10",
                   all(0 <= c["ExperienceYears"] <= 10 for c in candidates)))

# Skills validation
validation.append(("Minimum 3 Skills",
                   all(len(c["Skills"]) >= 3 for c in candidates)))

# JSON validation
try:
    json.dumps(candidates)
    json_ok = True
except:
    json_ok = False

validation.append(("Valid JSON", json_ok))

# Table validation
table_ok = not df.empty
validation.append(("Proper Table Format", table_ok))

errors = []

for check, status in validation:
    print(f"{check:35} : {'PASS' if status else 'FAIL'}")
    if not status:
        errors.append(check)

print("\nValidation Errors Identified:")

if len(errors) == 0:
    print("No validation errors found.")
else:
    for e in errors:
        print("-", e)

print("\n" + "="*80)
print("TASK 4: REFINED PROMPT")
print("="*80)

refined_prompt = """
Generate exactly 5 UNIQUE candidate profiles for an HR Dashboard.

Strict Requirements:

1. Output ONLY valid JSON followed by ONE table.
2. JSON must contain no syntax errors.
3. Each candidate must include:
   - CandidateID
   - Name
   - Age
   - ExperienceYears
   - Skills
   - Education
   - Location
4. Age must be between 21 and 35.
5. ExperienceYears must be between 0 and 10.
6. Skills must contain at least 3 UNIQUE skills.
7. CandidateID values must be unique.
8. Table must exactly match the JSON output.
9. Do not include any extra explanation.
"""

print(refined_prompt)

print("\n" + "="*80)
print("FINAL CORRECTED OUTPUT")
print("="*80)

print("JSON:\n")
print(json.dumps(candidates, indent=4))

print("\nTABLE:\n")
print(df.to_string(index=False))

print("\n" + "="*80)
print("FINAL VALIDATION SUMMARY")
print("="*80)

for check, status in validation:
    print(f"{check:35} : {'PASS' if status else 'FAIL'}")

print("\nAssignment Completed Successfully.")
```
