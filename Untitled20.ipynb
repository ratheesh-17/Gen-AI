```python
# ===========================
# Prompt Engineering Assignment
# ===========================

from google import genai
from google.colab import userdata
import json

# ---------------------------
# Step 1: Connect to Gemini
# ---------------------------
api_key = userdata.get("api_key")
client = genai.Client(api_key=api_key)

# ---------------------------
# Task 1: Problem Identification
# ---------------------------
print("="*70)
print("TASK 1: PROBLEM IDENTIFICATION")
print("="*70)

original_prompt = "Generate employee details"

print("\nOriginal Prompt:")
print(original_prompt)

problems = [
    "1. The prompt is too vague and lacks clarity.",
    "2. No output format is specified.",
    "3. Required fields are not mentioned.",
    "4. Number of employee records is not specified.",
    "5. No validation constraints for salary."
]

print("\nProblems Identified:")
for problem in problems:
    print(problem)

# ---------------------------
# Task 2: Improved Prompt
# ---------------------------
print("\n" + "="*70)
print("TASK 2: IMPROVED PROMPT")
print("="*70)

improved_prompt = """
Generate fictional employee details.

Requirements:
1. Return the output only in valid JSON format.
2. Generate exactly 5 employee records.
3. Each employee must contain:
   - name
   - department
   - salary
4. Salary must be numeric only.
5. Do not include explanations or markdown.
"""

print(improved_prompt)

# ---------------------------
# Task 3: Generate Output
# ---------------------------
print("\n" + "="*70)
print("TASK 3: FIRST GENERATED OUTPUT")
print("="*70)

response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents=improved_prompt
)

generated_output = response.text

print(generated_output)

# ---------------------------
# Task 4: Validation
# ---------------------------
print("\n" + "="*70)
print("TASK 4: OUTPUT VALIDATION")
print("="*70)

errors = []

try:
    employees = json.loads(generated_output)

    print("✅ JSON Format: Valid")

    if len(employees) != 5:
        errors.append("Exactly 5 records are required.")

    required_fields = {"name", "department", "salary"}

    for i, emp in enumerate(employees, start=1):

        if not required_fields.issubset(emp.keys()):
            errors.append(f"Employee {i} is missing required fields.")

        if not isinstance(emp["salary"], (int, float)):
            errors.append(f"Employee {i} salary is not numeric.")

except Exception as e:
    errors.append("Invalid JSON Format")

if len(errors) == 0:
    print("✅ All required fields are present.")
    print("✅ Exactly 5 records generated.")
    print("✅ Salary values are numeric.")
else:
    print("Errors Found:")
    for error in errors:
        print("-", error)

# ---------------------------
# Task 5: Refined Prompt
# ---------------------------
print("\n" + "="*70)
print("TASK 5: REFINED PROMPT")
print("="*70)

refined_prompt = """
Generate exactly 5 fictional employee records.

Return ONLY a valid JSON array.

Each object must contain:

{
"name": string,
"department": string,
"salary": number
}

Rules:
- Exactly 5 records
- Salary must be numeric only
- No markdown
- No explanation
- Output only JSON
"""

print(refined_prompt)

# ---------------------------
# Final Output
# ---------------------------
print("\n" + "="*70)
print("FINAL CORRECTED OUTPUT")
print("="*70)

final_response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents=refined_prompt
)

final_output = final_response.text

print(final_output)

# ---------------------------
# Final Validation
# ---------------------------
print("\n" + "="*70)
print("FINAL VALIDATION")
print("="*70)

try:
    employees = json.loads(final_output)

    print("✅ JSON Format: Valid")
    print("✅ Number of Records:", len(employees))

    required_fields = {"name", "department", "salary"}

    valid = True

    for emp in employees:
        if not required_fields.issubset(emp.keys()):
            valid = False
        if not isinstance(emp["salary"], (int, float)):
            valid = False

    if valid and len(employees) == 5:
        print("✅ All required fields are present.")
        print("✅ Salary is numeric.")
        print("✅ Output satisfies all constraints.")
    else:
        print("❌ Validation Failed.")

except Exception as e:
    print("❌ Invalid JSON")
    print(e)
```
