```python
from google import genai
from google.colab import userdata

# API Key
api_key = userdata.get("api_key")
client = genai.Client(api_key=api_key)

prompt = """
Scenario:
You are working as a Project Manager responsible for communicating updates after a sprint review.
The project has encountered:
- One-week delay
- Resource shortages
- Additional client feature requests

Task 1:
Create:
1. Professional Email (Maximum 120 words)
2. Structured Report

Report must contain:
- Title
- Summary
- Issues
- Action Plan

Task 2:
Evaluate whether:
- Email is professional
- Report contains all required sections
- Mention any missing content

Task 3:
Improve the prompt for better clarity.

Task 4:
Regenerate the email and report using the improved prompt.

Return everything in this order:

1. Initial Prompt
2. First Output (Email + Report)
3. Evaluation
4. Improved Prompt
5. Final Output
"""

response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents=prompt
)

print(response.text)
```
